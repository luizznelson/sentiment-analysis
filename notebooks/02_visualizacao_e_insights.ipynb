```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/reviews_sentimentos.csv")
```

```python
# 1. Distribuição dos sentimentos
plt.figure(figsize=(6,4))
sns.countplot(x="sentimento", data=df, palette="RdYlGn")
plt.title("Distribuição dos Sentimentos")
plt.xlabel("Sentimento")
plt.ylabel("Nº de reviews")
plt.savefig("../imgs/sentimento_distribuicao.png")
plt.show()
```

```python
# 2. Sentimento por serviço
plt.figure(figsize=(10,6))
sns.countplot(x="servico", hue="sentimento", data=df, palette="RdYlGn")
plt.title("Sentimento por Serviço")
plt.xlabel("Serviço")
plt.ylabel("Nº de reviews")
plt.legend(title="Sentimento")
plt.savefig("../imgs/sentimento_por_servico.png")
plt.show()
```

```python
# 3. Top clientes mais engajados (que mais avaliam)
top_clientes = df['cliente'].value_counts().head(10)
plt.figure(figsize=(8,4))
sns.barplot(x=top_clientes.values, y=top_clientes.index, palette="Blues_r")
plt.title("Clientes com Mais Reviews")
plt.xlabel("Nº de reviews")
plt.ylabel("Cliente")
plt.savefig("../imgs/top_clientes.png")
plt.show()
```

```python
# 4. Exemplo: nuvem de palavras para reviews negativos
from wordcloud import WordCloud

text_neg = " ".join(df[df["sentimento"]=="Negativo"]["review"].values)
wc = WordCloud(width=600, height=300, background_color='white', colormap='Reds').generate(text_neg)
plt.figure(figsize=(10,5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("Nuvem de Palavras - Reviews Negativos")
plt.savefig("../imgs/nuvem_negativo.png")
plt.show()
```
